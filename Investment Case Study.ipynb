```python
# Write your code for importing the csv file here
import csv
#importing companies file
companies=pd.read_csv("https://cdn.upgrad.com/UpGrad/temp/d934844e-5182-4b58-b896-4ba2a499aa57/companies.txt", sep="\t", encoding = "ISO-8859-1")
companies
```

```python
#importing rounds2 csv file
rounds2=pd.read_csv("https://cdn.upgrad.com/UpGrad/temp/4c3b5ed0-e5dc-4838-89a2-173d8707d857/rounds2.csv", encoding = "ISO-8859-1")
rounds2
```

```python
#Handling encoding issues in companies dataframe
companies['permalink'] = companies[['permalink']].applymap(lambda x: x.encode('utf-8').decode('ascii', 'ignore'))
companies
```

```python
#handling encoding issues in rounds2 dataframe
rounds2['company_permalink'] = rounds2[['company_permalink']].applymap(lambda x: x.encode('utf-8').decode('ascii', 'ignore'))
rounds2
```

```python
# no. of unique companies present in rounds2
len(rounds2['company_permalink'].unique())
```

```python
#no. of unique companies present in companies
len(companies['permalink'].unique())
```

```python
companies['permalink']=companies['permalink'].str.lower()
rounds2['company_permalink']=rounds2['company_permalink'].str.lower()
```

```python
#companies['permalink'].isin(rounds2['company_permalink'])
#len(rounds2['company_permalink'].isin(companies['permalink']).unique())
#companies['permalink'].isin(rounds2['company_permalink'])
len(set(companies['permalink']).difference(rounds2['company_permalink']))
```

```python
master_frame = pd.merge(rounds2, companies, left_on='company_permalink', right_on='permalink',how='left')
master_frame
```

```python
master_frame.drop_duplicates(keep = False, inplace = True) 
len(master_frame)                     
```

```python
round(100*(master_frame.isnull().sum()/len(master_frame.index)),2)
```

```python
master_frame=master_frame[master_frame.isnull().sum(axis=1) <=5]
round(100*(master_frame.isnull().sum()/len(master_frame.index)),2)
```

```python
master_frame=master_frame[~np.isnan(master_frame['raised_amount_usd'])]

master_frame=master_frame[~pd.isnull(master_frame['country_code'])]

round(100*(master_frame.isnull().sum()/len(master_frame.index)),2)
```

```python
master_frame=master_frame.drop(['funding_round_code','homepage_url','founded_at','state_code','region','city'],axis=1)
```

```python
round(100*(master_frame.isnull().sum()/len(master_frame.index)),2)
```

```python
#Converting the unit of the raised_amount_usd column from $ to million $.
master_frame['raised_amount_usd']=master_frame['raised_amount_usd'].apply(lambda x : round(x/1000000,2))
master_frame
```

```python
investment_group_by=master_frame.groupby(['funding_round_type'])
investment_group_by
```

```python
#Average Values of Investments for Each Funding Types 
Avg_investment=pd.DataFrame(investment_group_by.raised_amount_usd.mean().sort_values(ascending=False))
Avg_investment
```

```python
# grouping by Country 
investment_type=pd.DataFrame(master_frame.loc[master_frame.funding_round_type == 'venture'])
group_by_country=investment_type.groupby('country_code')
group_by_country

```

```python
#countries with the highest amount of funding for the chosen investment type(venture)
top9=pd.DataFrame(group_by_country.raised_amount_usd.sum().sort_values(ascending=False)).head(9)
top9
```

```python
#importing mapping.csv file
mapping=pd.read_csv("https://cdn.upgrad.com/UpGrad/temp/231dc91c-0642-470d-a362-29ddcd7142ce/mapping.csv", encoding = "ISO-8859-1")
mapping
```

```python
# code for extracting the first string from category_list column
mapping['category_list']=mapping.category_list.str.split("|").str[0]
mapping
```

```python
# code to Extract the primary sector of each category list from the category_list column
mapping=pd.melt(mapping,id_vars=['category_list'],value_vars=['Automotive & Sports','Cleantech / Semiconductors',
            'Entertainment','Health','Manufacturing','News, Search and Messaging','Others','Social, Finance, Analytics, Advertising',
             ],var_name='main_sector'
            )
mapping=mapping[(mapping.value==1)]
mapping
```

```python
#Replacing all the zeros with n 

mapping['category_list']=[x.strip().replace('0', 'na') for x in mapping['category_list']]
mapping['category_list']=[x.strip().replace('.na', '0') for x in mapping['category_list']]
mapping
```

```python
#checkpoint 5
#Mapping catagory list to main_catagory in master dataframe
master_frame=pd.DataFrame(master_frame.merge(mapping,on='category_list'))
master_frame
```

```python
#creating dataframe D1 for 1st english speaking country USA and funding type venture
D1= pd.DataFrame(master_frame.loc[(master_frame.country_code=='USA')&(master_frame.funding_round_type=='venture')])
D1
```
